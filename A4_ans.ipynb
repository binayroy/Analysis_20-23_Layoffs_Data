### Question - 1

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
city = pd.read_html('assets/wikipedia_data.html')[1]
city.rename(columns={'Population (2016 est.)[8]':'Population',
                    'Metropolitan area':'Metro'},inplace=True)
city = city.iloc[0:-1,[0,3,5,6,7,8]]
city = city.replace(to_replace='\[.*?\]',value='',regex=True)
citynhl = city[['Metro','Population','NHL']]
citynhl.iloc[5,0],citynhl.iloc[0,0] = 'Washington','New York'
citynhl = citynhl.set_index(['Metro','Population'])
citynhl = citynhl.apply(lambda x: x.str.split(' ').explode())
citynhl = citynhl.reset_index()
citynhl['NHL'] = citynhl['NHL'].replace('—',np.nan).replace('',np.nan)
citynhl.dropna(inplace=True)
#print(citynhl.shape)
citynhl
```

```python
nhldf = pd.read_csv('assets/nhl.csv')
nhldf = nhldf[['team','W','L','year']]
nhldf = nhldf[nhldf['year']==2018]
nhldf = nhldf[nhldf['L'] != nhldf['W']]
nhldf['W'],nhldf["L"] = nhldf['W'].apply(int),nhldf["L"].apply(int)
nhldf['W/L'] = nhldf['W']/(nhldf['W']+nhldf['L'])
nhldf = nhldf[['team','W/L']]
nhldf.replace(to_replace='\*',value='',regex=True,inplace=True)
nhldf['NHL'] = nhldf['team'].str.split().str.get(-1)
nhldf
```

```python
df = pd.merge(nhldf,citynhl, how='inner')
df['Population'] = df['Population'].apply(int)
df = df.groupby('Metro').agg({'W/L': np.nanmean, 'Population': np.nanmean})
df.shape
df
```

```python
stats.pearsonr(df['Population'], df['W/L'])[0]
```

### NFL

```python
# Wikipedia Data:
city = pd.read_html('assets/wikipedia_data.html')[1]
city = city.rename({'Metropolitan area':'Metro','Population (2016 est.)[8]': 'Population'},axis=1)
city = city[:-1]

# NFL
citynfl = city[['Metro', 'Population','NFL']]
citynfl = citynfl.replace('\[.*','',regex=True).replace(['— ','—',''],[np.nan,np.nan,np.nan])
citynfl.dropna(inplace=True)
citynfl.set_index(['Metro', 'Population'],inplace=True)
citynfl = citynfl.apply(lambda x: x.str.split(' ').explode()).reset_index()
citynfl

nfldf = pd.read_csv('assets/nfl.csv')
nfldf = nfldf[nfldf['year']==2018]
nfldf = nfldf[['team','W-L%']]
nfldf.columns=['teams','WL']
nfldf = nfldf[nfldf['teams'] != nfldf['WL']]
nfldf = nfldf.replace('\*.*','',regex=True).replace('\+.*','',regex=True)
nfldf['NFL'] = nfldf['teams'].apply(lambda x: x.split(' ')[-1])
nfldf

# Merging
nfldf = pd.merge(citynfl,nfldf,how='outer')
nfldf = nfldf.astype({'Population': int,'WL': float})
nfldf = nfldf.groupby('Metro').agg({'WL':np.nanmean})
nfldf.columns = ['NFL']
nfldf = nfldf.reset_index()
nfldf
```
